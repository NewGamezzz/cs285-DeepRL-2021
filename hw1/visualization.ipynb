```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Question 1.2: Behavior Cloning

## Number of training steps per iteration

### Ant Environment

```python
mean_return = [1739.3922, 1768.6057, 2156.6765, 2108.8867, 3635.514, 3820.3354, 3707.5366]
std_return = [1035.5768, 1104.6294, 1380.1221, 1596.4115, 846.1677, 1411.9393, 780.2101]

train_step = [1000, 1250, 1500, 1750, 2000, 2250, 2500]

expert_return = [4679.166] * len(train_step)

plt.errorbar(train_step, mean_return, yerr=std_return, label="Trained Policy")
plt.errorbar(train_step, expert_return, label="Expert Policy")
plt.xlabel("Number of Training Step per Iteration")
plt.ylabel("Mean & Std of Return")
plt.title("The relationship between training steps and Return (Ant-v2)")
plt.legend()
plt.show()
```

# Question 2.2: DAgger

## Learning Curve

### Ant Environment

```python
mean_return = pd.read_csv("data/q2_dagger_ant_Ant-v2_01-01-2023_16-28-14/mean_return.csv")
std_return = pd.read_csv("data/q2_dagger_ant_Ant-v2_01-01-2023_16-28-14/std_return.csv")

mean_return = mean_return["Value"].values
std_return = std_return["Value"].values

expert_return = [4679.166] * len(mean_return)

bc_mean_return = [1739.3922] * len(mean_return)
bc_std_return = [1035.5768] * len(mean_return)

plt.errorbar(np.arange(len(mean_return)), mean_return, yerr=std_return, label="DAgger")
plt.errorbar(np.arange(len(mean_return)), expert_return, label="Expert Policy")
plt.errorbar(np.arange(len(mean_return)), bc_mean_return, yerr=bc_std_return, label="Behavior Cloning Policy")
plt.xlabel("Iteration")
plt.ylabel("Mean & Std of Return")
plt.title("Learning Curve (Ant-v2)")
plt.legend()
plt.show()
```

### Humanoid Environment

```python
mean_return = pd.read_csv("data/q2_dagger_humanoid_Humanoid-v2_01-01-2023_17-24-36/mean_return.csv")
std_return = pd.read_csv("data/q2_dagger_humanoid_Humanoid-v2_01-01-2023_17-24-36/std_return.csv")

mean_return = mean_return["Value"].values
std_return = std_return["Value"].values

expert_return = [10714.461] * len(mean_return)

bc_mean_return = [282.97147] * len(mean_return)
bc_std_return = [66.82996] * len(mean_return)

plt.errorbar(np.arange(len(mean_return)), mean_return, yerr=std_return, label="DAgger")
plt.errorbar(np.arange(len(mean_return)), expert_return, label="Expert Policy")
plt.errorbar(np.arange(len(mean_return)), bc_mean_return, yerr=bc_std_return, label="Behavior Cloning Policy")
plt.xlabel("Iteration")
plt.ylabel("Mean & Std of Return")
plt.title("Learning Curve (Humanoid-v2)")
plt.legend()
plt.show()
```
